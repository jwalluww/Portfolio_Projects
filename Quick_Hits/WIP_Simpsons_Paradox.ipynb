# Simpsons Paradox
- How the overall trend can reverse when the data is stratified across two groups

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Sample sizes
n_new_old = 800
n_return_old = 200

n_new_new = 200
n_return_new = 800

# Conversion rates (new onboarding is better for BOTH groups)
conv_new_old = 0.05
conv_new_new = 0.06

conv_return_old = 0.30
conv_return_new = 0.32

data = []

# Old onboarding
data += [("old", "new_user", np.random.binomial(1, conv_new_old)) for _ in range(n_new_old)]
data += [("old", "returning_user", np.random.binomial(1, conv_return_old)) for _ in range(n_return_old)]

# New onboarding
data += [("new", "new_user", np.random.binomial(1, conv_new_new)) for _ in range(n_new_new)]
data += [("new", "returning_user", np.random.binomial(1, conv_return_new)) for _ in range(n_return_new)]

df = pd.DataFrame(data, columns=["onboarding", "user_type", "converted"])
```

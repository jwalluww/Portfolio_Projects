# Step 1: Gather Data
- Preferred format: .txt
- Topic ASOIAF Wiki

```python
import requests

def fetch_awioaf_page(title):
    endpoint = "https://awoiaf.westeros.org/api.php"
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "explaintext": True,
        "titles": title
    }
    response = requests.get(endpoint, params=params)
    pages = response.json()["query"]["pages"]
    return next(iter(pages.values()))["extract"]

# Example:
print(fetch_awioaf_page("Jon Snow"))
```
